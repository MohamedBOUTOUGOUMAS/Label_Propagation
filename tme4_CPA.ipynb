```python
import networkx as nx
import random as rn
import matplotlib.pyplot as plt
```

# Exercice1

```python
def constructG(G,p,q,nbNode,nameFile):
    open(nameFile +".data", 'w').close()
    f = open(nameFile + ".data", "a")
    for i in range(nbNode):
        for j in range(i+1,nbNode):
            if (i // (nbNode//4)) == (j // (nbNode//4)):
                if rn.random() <= p:
                    G.add_edge(i, j)
                    f.write(str(i) + " " + str(j) + "\n")
                    f.write(str(j) + " " + str(i) + "\n")
            else :
                if rn.random() <= q:
                    G.add_edge(i, j)
                    f.write(str(i) + " " + str(j) + "\n")
                    f.write(str(j) + " " + str(i) + "\n")
    f.close()
```

```python
nbNode = 400
p = 0.25
q = 0.0
G = nx.Graph()
G.add_nodes_from(list(range(nbNode)))
constructG(G,p,q,nbNode,"graphe1")
nx.draw_spring(G)
plt.draw()
```
